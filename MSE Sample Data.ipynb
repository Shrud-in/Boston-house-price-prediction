```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
```

```python
# generating evenly spaced values froom 0 to 2 PI radians
n=200
x=np.linspace(0,2*np.pi,n)
sine_values=np.sin(x)
plt.plot(x,sine_values,color='b')
```

```python
# Adding noise to sine wave
noise=0.5
noisy_sine_values=sine_values+np.random.uniform(-noise,noise,n)

#Plotting the noisy data
plt.plot(x,noisy_sine_values, color='r')
plt.plot(x,sine_values, linewidth=5)
```

```python
# Calculating MSE manually using the traditional formula
error_value=(1/n)*sum(np.power(sine_values - noisy_sine_values,2))
error_value
```

```python
# Calculating MSE using function from sklearn library
mean_squared_error(sine_values,noisy_sine_values)
```
